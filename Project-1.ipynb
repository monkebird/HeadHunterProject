1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
# Функция-преобразование для получения категории образования
def education_level(text):
    words = text.split()
    if words[0] == 'Высшее':
        return 'высшее'
    elif words[0] == 'Неоконченное':
        return 'неоконченное высшее'
    elif words[0] == 'Среднее' and words[1] == 'специальное':
        return 'среднее специальное'
    elif words[0] == 'Среднее':
        return 'среднее'
    else:
        return 'неизвестно'
    
# Создаем новый столбец "Образование" с помощью функции-преобразования
hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(lambda x: education_level(x))

# Удаляем столбец "Образование и ВУЗ"
hh_data.drop('Образование и ВУЗ', axis=1, inplace=True)

# 3.1. Сколько соискателей имеет средний уровень образования (школьное образование)?
print(hh_data['Образование'].value_counts()['среднее'])
```

* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
# Функция-преобразование для пола
def gender(text):
    gender = text.split(',')[0].strip()
    if gender == 'Мужчина':
        return 'М'
    else:
        return 'Ж'

# Функция-преобразование для возраста
def age(text):
    age = text.split(',')[1].strip().split()[0]
    return int(age)

# Создаем новые столбцы "Пол" и "Возраст" с помощью функций-преобразований
hh_data['Пол'] = hh_data['Пол, возраст'].apply(lambda x: gender(x))
hh_data['Возраст'] = hh_data['Пол, возраст'].apply(lambda x: age(x))

# Удаляем столбец "Пол, возраст"
hh_data.drop('Пол, возраст', axis=1, inplace=True)

# 3.2. Сколько процентов женских резюме представлено в наших данных? Чему равен средний возраст соискателей? (Ответ округлите до сотых)
print(round(hh_data['Пол'].value_counts(normalize=True)['Ж'] * 100, 2))
print(round(hh_data['Возраст'].mean(), 2))
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Функция-преобразование для опыта работы
def extract_experience(text):
    
    # Если пропуск или "Не указано", возвращаем NaN
    if pd.isna(text) or text == 'Не указано':
        return np.nan
    
    else:
        # Разбиваем текст опыта работы на слова
        words = text.split(' ')[:7]

        # Инициализируем переменные для годов и месяцев
        years = 0
        months = 0
        total_months = 0
        
        # Создаем списки ключевых слов, соответствующих годам и месяцам
        month_list= ['месяц', 'месяца', 'месяцев']
        year_list = ['год', 'года', 'лет']
        
        # Проходим по словам и ищем ключевые слова
        for i, item in enumerate(words):
            if item in year_list:
                years = int(words[i - 1])
            elif item in month_list:
                months = int(words[i - 1])
                
         # Переводим опыт работы в месяцы и возвращаем
        total_months = (years * 12) + months
        return total_months

# Применяем функцию-преобразование к столбцу "Опыт работы" 
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(lambda x: extract_experience(x))

# Удаляем столбец "Опыт работы" из таблицы
hh_data.drop('Опыт работы', axis=1, inplace=True)

# 3.3. Чему равен медианный опыт работы (в месяцах) в нашей таблице?
print(round(hh_data['Опыт работы (месяц)'].median()))
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

```python
# Функция-преобразование для города
def get_city(text):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород',
                      'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 
                      'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
                     ]
    city = text.split(' , ')[0]
    if (city == 'Москва') or (city == 'Санкт-Петербург'):
        return city
    elif city in million_cities:
        return 'город миллионник'
    else:
        return 'другие'

# Извлекаем информацию о готовности к переезду    
def get_ready_to_move(text):
    if ('не готов к переезду' in text) or ('не готова к переезду' in text):
        return False
    elif 'хочу' in text:
        return True
    else:
        return True

# Функция-преобразование для командировок
def get_ready_for_bisiness_trips(text):
    if ('командировка' in text):
        if ('не готов к командировкам' in text) or('не готова к командировкам' in text):
            return False
        else: 
            return True
    else:
        return False

# Применяем функции к каждойму признаку    
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(get_city)
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(get_ready_to_move)
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(get_ready_for_bisiness_trips)

# Удаляем столбец "Город, переезд, командировки" из таблицы    
hh_data.drop('Город, переезд, командировки', axis=1, inplace=True)

# 3.4. Сколько процентов соискателей живут в Санкт-Петербурге? Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам?
print(round(hh_data['Город'].value_counts(normalize=True)['Санкт-Петербург'] * 100)) 
print(round(hh_data[hh_data['Готовность к переезду'] & hh_data['Готовность к командировкам']].shape[0] / hh_data.shape[0] *100))
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:

<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>

Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Создаем признаки-мигалки
employment_categories = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
schedule_categories = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

for category in employment_categories:
    hh_data[category] = hh_data['Занятость'].apply(lambda x: category in x)

for category in schedule_categories:
    hh_data[category] = hh_data['График'].apply(lambda x: category in x)

# Удаляем исходные признаки "Занятость" и "График" 
hh_data.drop(['Занятость', 'График'], axis=1, inplace=True)
    
# 3.5. Сколько людей ищут проектную работу и волонтёрство? Сколько людей хотят работать вахтовым методом и с гибким графиком?
print(hh_data[hh_data['проектная работа'] & hh_data['волонтерство']].shape[0])
print(hh_data[hh_data['вахтовый метод'] & hh_data['гибкий график']].shape[0])
```

Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Открываем файл с данными о курсе валют
exchange_rates = pd.read_csv('data/ExchangeRates.csv', sep=',')

# Приводим признаки "Обновление резюме" и "date" в формат datetime
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True)
exchange_rates['date'] = pd.to_datetime(exchange_rates['date'], format='%d/%m/%y', dayfirst=True)

# Удаляем ненужные столбцы
exchange_rates.drop(columns=['per', 'time', 'vol'], axis=1, inplace=True)

# Достаем дату из признака "Обновление резюме"
hh_data['Обновление резюме'] = hh_data['Обновление резюме'].dt.date

# Выделяем из столбца "ЗП" сумму желаемой заработной платы и наименование валюты
hh_sal = pd.DataFrame()
hh_sal['Зарплата'] = hh_data['ЗП'].apply(lambda x: str(x).split()[0]).astype(float)
hh_sal['currency'] = hh_data['ЗП'].apply(lambda x: str(x).split()[1])
hh_sal['date'] = pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True)

# Переводим наименования валют в стандарт ISO
hh_sal['currency'].replace({'грн.':'UAH', 'руб.':'RUB', 'бел.руб.':'BYN', 'сум':'UZS'}, inplace=True)

# Объединяем таблицы
hh_sal = pd.merge(hh_sal, exchange_rates, on=['date', 'currency'], how='left')

# Заполняем пропуски для рубля единицей
hh_sal['close'].fillna(1, inplace=True)
hh_sal['proportion'].fillna(1, inplace=True)

# Создаем столбец "ЗП (руб)", куда вносим конвертированную сумму
hh_data['ЗП (руб)'] = ((hh_sal['Зарплата'] * hh_sal['close']) / hh_sal['proportion']).astype(int)

# 3.6. Чему равна желаемая медианная заработная плата (в рублях)?
print(round(hh_data['ЗП (руб)'].median()/1000))
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Создаем объекты для рисования графика
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

# Применяем логарифмическое преобразование к столбцу "Возраст" + 1
log_age = np.log(hh_data['Возраст'] + 1)

# Строим гистограмму в логарифмическом масштабе с 30 бинами
histplot = sns.histplot(log_age, bins=30, ax=ax)

# Добавляем вертикальные линии для среднего значения, верхней и нижней границы 4 стандартных отклонений
histplot.axvline(log_age.mean(), color='k', lw=2) # Среднее значение
histplot.axvline(log_age.mean() + 4 * log_age.std(), color='k', ls='--', lw=2) # Верхняя граница 4 сигмы
histplot.axvline(log_age.mean() - 3 * log_age.std(), color='k', ls='--', lw=2) # Нижняя граница 3 сигмы

# Устанавливаем заголовок для графика
histplot.set_title('Распределение возраста в логорифмическом масштабе')

# Функция для обработки выбросов с использованием Z-отклонения
def outliers_z_score_mod(data, feature, left=3, right=3, bias=1, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+bias)
    else:
        x = data[feature]
    
    mu = x.mean() 
    sigma = x.std() 
    
    lower_bound = mu - left * sigma 
    upper_bound = mu + right * sigma 
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(hh_data, 'Возраст', 3, 4, bias=0, log_scale=True)

# Вызываем функцию для поиска выбросов в столбце "Возраст" с логарифмическим масштабом
outliers, cleaned_data = outliers_z_score_mod(hh_data, 'Возраст', left=3, right=4, bias=0, log_scale=True)

# Выводим количество найденных выбросов
print(outliers.shape[0])

# Выводим таблицу с выбросами
print("Выбросы в возрасте (логарифмический масштаб):")
print(outliers[['Возраст', 'ЗП (руб)']])
```

График показывает, что распределение ассиметрично вправо. Это означает, что большинство значений сконцентрированы в низких значениях, а хвост распределения идет вправо к более высоким значениям возраста.
